```python
from sklearn.neighbors import KNeighborsClassifier 
import pandas as pd 
df=pd.read_csv("mobile_price.csv")
df
```

```python
x=df[["ram" ,"storage" ,"battery"]]
y=df["price_range"]
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x,y)
ram=int(input("Enter your ram : "))
storage=int(input("Enter your storage : "))
battery=int(input("Enter your battery : "))
new_data=pd.DataFrame([[ram ,storage ,battery]] ,columns=["ram" ,"storage" ,"battery"])
pred=model.predict(new_data)[0]
if pred == 0:
    print("📱 Low budget phone")
elif pred == 1:
    print("📱 Medium range phone")
else:
    print("📱 High-end phone 🔥")
```
